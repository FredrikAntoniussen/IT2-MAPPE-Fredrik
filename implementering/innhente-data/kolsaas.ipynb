```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

fil = open("kolsaas.csv")
data = fil.read()
```

```python
linjer = data.split("\n")
tur = []
for linje in linjer:
    splittet_linje = linje.split(",")
    tur.append(splittet_linje)

```

```python
#tid x-akse mot høyde 
lengde = []
bredde = []
høyde = []
x =[]
y = []

for linje in tur[1:]:
    høyde = float(linje[2])
    lengde.append(float(linje[2]))
    bredde.append(float(linje[1]))
    tid = (float(linje[0]))
    x.append(tid)
    y.append(høyde)

    
  
    


```
