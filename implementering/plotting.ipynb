```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

x = []

y = []

def f(x):
    return 3*x**2
for i in range(6):
    x.append(i)
    y.append(f(i))
    
plt.scatter(x,y)
plt.show()
```

```python
x = np.linspace(0,200,10)
y = f(x)
```

```python
plt.plot(x,y)
plt.title("Title")
plt.xlabel("$x$")
plt.ylabel("$y$")

plt.show()
```
